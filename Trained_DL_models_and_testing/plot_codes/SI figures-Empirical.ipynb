```python
import os
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt 
import sys
import csv
```

```python
model = 'COVID'

df_ews = pd.read_csv('../data/ews/df_ews_forced_{}.csv'.format(model)) 

for i in range(1,8): 

    df_ml = pd.read_csv('../data/ml_pred_SIR_CL100/ensemble_trend_probs_{}forced_{}_len100.csv'.format(model, i))
    df = df_ml.iloc[:,0:2]

    plot_num = i
    var = 'I'

    df_plot_num = df_ews[(df_ews['tsid'] == plot_num)]
    tbif = df_plot_num[['Time','residuals']].dropna()['Time'].iloc[-1]

    post_transition_tsid = 4
    tbif1 = tbif + (post_transition_tsid-4)*28 

    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(6, 6))

    color = 'tab:blue' 
    df_I = df_ews.loc[df_ews['Variable'] == var]
    t = df_I.loc[df_ews['tsid'] == plot_num][['Time']]
    ax1.plot(t, df_I.loc[df_ews['tsid'] == plot_num][['state','smoothing']], label=['state','smoothing'])
    ax1.set_xlabel('time (days)', fontsize = 16)
    ax1.set_ylabel('Infected Individuals', fontsize = 16)
    ax1.axvline(tbif, color ='r', linestyle = 'dashed', label = 'Transition Point')
    ax1.legend(fontsize = 14)
    ax1.tick_params(axis='y', labelsize=14)
    ax1.tick_params(axis='x', labelsize=14)
    
    gap = 10
    time_points = range(t['Time'].max(), -1, -gap)
    filtered_df = t[t['Time'].isin(time_points)]

    t1 = filtered_df.reset_index(drop=True)
    
    prediction = df_ml.tail(len(t1))

    ax2.scatter(t1, prediction.iloc[:len(t1), 0:1], color='tab:blue', label='prob. for Null')  # Adjusted y-axis to match the size of t1
    ax2.scatter(t1, prediction.iloc[:len(t1), 1:2], color='tab:orange', label='prob. for Transcritical')  # Adjusted y-axis to match the size of t1

    ax2.legend()
    ax2.set_ylabel('Probability', fontsize = 16)
    ax2.set_xlabel('time (days)', fontsize = 16)
    ax2.legend(fontsize = 14)
    ax2.set_ylim(-0.2,1.2)
    ax2.tick_params(axis='y', labelsize=14)
    ax2.tick_params(axis='x', labelsize=14)
    
    plt.tight_layout()
    plt.subplots_adjust(hspace=0)

    plt.savefig('../data/figures/SI_figures/{}_forced_plot{}.png'.format(model, i))

```
