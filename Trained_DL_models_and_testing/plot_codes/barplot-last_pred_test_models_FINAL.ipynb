```python
'''
plot of the last predictions by SIR trained model on four test models.
Inputs are the ensemble prediction files of the DL model and outputs 
are the bar plot used in fig 2.
'''

import pandas as pd 
import matplotlib.pyplot as plt 

test_models = ['SIRwhiteN', 'SIRenvN', 'SIRdemN', 'SEIR']
types = ['forced', 'null']
new_columns = ['null_prob', 'branch_prob', 'bif_prob']
 
NoOFfiles = 10
df_full = pd.DataFrame()
df_ind_model = pd.DataFrame()

for test_model in test_models:    
    df_ind_model = pd.DataFrame()
    for type in types: 
        for i in range(1, NoOFfiles+1): 
            df_ml = pd.read_csv(r'../data/ml_pred_SIR/ensemble_trend_probs_{}{}_{}_len500.csv'.format(test_model, type, i)) 
            df_temp = df_ml.iloc[-1].to_frame().T
            df_temp.columns = new_columns
            df_temp['Simulation_No'] = test_model + type + '_' + str(i)
            df_ind_model = pd.concat([df_ind_model, df_temp], ignore_index=True)
            
    df_full = pd.concat([df_full, df_ind_model], axis=1)


# print('last prediction from each ensembled file:', df_full)


#For white noise model
df_whiteN = df_full.iloc[:, [0, 1, 2, 3]]

# creating dataframe 
df = pd.DataFrame({ 
	'Simulations': [str(i) for i in range(1, NoOFfiles+1)], 
	'Null': df_whiteN['null_prob'][0:10], 
	'Transcritical': df_whiteN['branch_prob'][0:10] 
}) 

df.plot(x="Simulations", y=["Null", "Transcritical"], kind="bar", figsize=(5, 2)) 
plt.xlabel('Simulations', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.ylim(0, 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)

plt.savefig(r'../data/figures/fig-2a1-whiteN-last-prob.png', bbox_inches = 'tight')

print('Last prediction among Transcritical simulations of White Noise model:')
plt.show()

# creating dataframe 
df = pd.DataFrame({ 
	'Simulations': [str(i) for i in range(1, NoOFfiles+1)], 
	'Null': df_whiteN['null_prob'][10:20], 
	'Transcritical': df_whiteN['branch_prob'][10:20] 
}) 

plt.figure(figsize=(10, 2))
df.plot(x="Simulations", y=["Null", "Transcritical"], kind="bar", figsize=(5, 2)) 
plt.xlabel('Simulations', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.ylim(0, 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)

plt.savefig(r'../data/figures/fig-2a2-whiteN-last-prob.png', bbox_inches = 'tight')

print('Last prediction among Null simulations of White Noise model:')
plt.show()


#for environmental noise model
df_envN = df_full.iloc[:, [4, 5, 6, 7]]
df_envN


df = pd.DataFrame({ 
	'Simulations': [str(i) for i in range(1, NoOFfiles+1)], 
	'Null': df_envN['null_prob'][0:10], 
	'Transcritical': df_envN['branch_prob'][0:10] 
}) 

plt.figure(figsize=(10, 2))
df.plot(x="Simulations", y=["Null", "Transcritical"], kind="bar", figsize=(5, 2)) 
plt.xlabel('Simulations', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.ylim(0, 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)

plt.savefig(r'../data/figures/fig-2b1-envN-last-prob.png', bbox_inches = 'tight')

print('Last prediction among forced simulations of Env Noise model:')
plt.show()


df = pd.DataFrame({ 
	'Simulations': [str(i) for i in range(1, NoOFfiles+1)], 
	'Null': df_envN['null_prob'][10:20], 
	'Transcritical': df_envN['branch_prob'][10:20] 
}) 

plt.figure(figsize=(10, 2))
df.plot(x="Simulations", y=["Null", "Transcritical"], kind="bar", figsize=(5, 2)) 
plt.xlabel('Simulations', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.ylim(0, 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)

plt.savefig(r'../data/figures/fig-2b2-envN-last-prob.png', bbox_inches = 'tight')

print('Last prediction among Null simulations of Env Noise model:')
plt.show()


# for demographic noise model
df_demN = df_full.iloc[:, [8, 9, 10, 11]]

df = pd.DataFrame({ 
	'Simulations': [str(i) for i in range(1, NoOFfiles+1)], 
	'Null': df_demN['null_prob'][0:10], 
	'Transcritical': df_demN['branch_prob'][0:10] 
}) 

plt.figure(figsize=(10, 2))
df.plot(x="Simulations", y=["Null", "Transcritical"], kind="bar", figsize=(5, 2)) 
plt.xlabel('Simulations', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.ylim(0, 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)

plt.savefig(r'../data/figures/fig-2c1-demN-last-prob.png', bbox_inches = 'tight')

print('Last prediction among Transcritical simulations of Dem Noise model:')
plt.show()

df = pd.DataFrame({ 
	'Simulations': [str(i) for i in range(1, NoOFfiles+1)], 
	'Null': df_demN['null_prob'][10:20], 
	'Transcritical': df_demN['branch_prob'][10:20] 
}) 

plt.figure(figsize=(10, 2))
df.plot(x="Simulations", y=["Null", "Transcritical"], kind="bar", figsize=(5, 2)) 
plt.xlabel('Simulations', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.ylim(0, 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)

plt.savefig(r'../data/figures/fig-2c2-demN-last-prob.png', bbox_inches = 'tight')

print('Last prediction among Null simulations of Dem Noise model:')
plt.show()

#for SEIR model
df_SEIR = df_full.iloc[:, [12, 13, 14, 15]]

df_envN

df = pd.DataFrame({ 
	'Simulations': [str(i) for i in range(1, NoOFfiles+1)], 
	'Null': df_SEIR['null_prob'][0:10], 
	'Transcritical': df_SEIR['branch_prob'][0:10] 
}) 

plt.figure(figsize=(10, 2))
df.plot(x="Simulations", y=["Null", "Transcritical"], kind="bar", figsize=(5, 2)) 
plt.xlabel('Simulations', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.ylim(0, 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)

plt.savefig(r'../data/figures/fig-2d1-SEIR-last-prob.png', bbox_inches = 'tight')

print('Last prediction among Transcritical simulations of SEIR model:')
plt.show()

df = pd.DataFrame({ 
	'Simulations': [str(i) for i in range(1, NoOFfiles+1)], 
	'Null': df_SEIR['null_prob'][10:20], 
	'Transcritical': df_SEIR['branch_prob'][10:20] 
}) 

plt.figure(figsize=(10, 2))
df.plot(x="Simulations", y=["Null", "Transcritical"], kind="bar", figsize=(5, 2)) 
plt.xlabel('Simulations', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.ylim(0, 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)

plt.savefig(r'../data/figures/fig-2d2-SEIR-last-prob.png', bbox_inches = 'tight')

print('Last prediction among Null simulations of SEIR model:')
plt.show()

```
