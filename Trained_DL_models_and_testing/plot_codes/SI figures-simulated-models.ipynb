```python
import os
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt 
import sys
import csv
```

```python
#model are: 'SIRwhiteN', 'SIRenvN', 'SIRdemN', 'SEIR'
model = 'SEIR'           

df_ews = pd.read_csv('../data/ews/df_ews_forced_{}.csv'.format(model)) 

for i in range(1,11): 

    df_ml = pd.read_csv('../data/ml_pred_SIR_CL500/ensemble_trend_probs_{}forced_{}_len500.csv'.format(model, i))
    df = df_ml.iloc[:,0:2]

    plot_num = i
    var = 'I'

    df_plot_num = df_ews[(df_ews['tsid'] == plot_num)]
    tbif = df_plot_num[['Time','residuals']].dropna()['Time'].iloc[-1]

    start_arrow = tbif - 500

    if start_arrow <0: 
        start_arrow = 0


    post_transition_tsid = 4
    tbif1 = tbif + (post_transition_tsid-4)*28

    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(6, 6))

    color = 'tab:blue' 
    df_I = df_ews.loc[df_ews['Variable'] == var]
    t = df_I.loc[df_ews['tsid'] == plot_num][['Time']]
    ax1.plot(t, df_I.loc[df_ews['tsid'] == plot_num][['state','smoothing']], label=['state','smoothing'])
    ax1.set_xlabel('time', fontsize = 16)
    ax1.set_ylabel('Infected Individuals', fontsize = 16)
    ax1.axvline(tbif, color ='r', linestyle = 'dashed', label = 'Transition Point')
    ax1.annotate('', xy=(start_arrow, 5), xytext=(tbif1+7, 5), arrowprops=dict(arrowstyle='<->', color='black'))
    ax1.axvspan(0, start_arrow, color = 'gainsboro')
    ax1.axvspan(tbif1, 1500, color = 'gainsboro')
    ax1.legend(fontsize = 14)
    ax1.tick_params(axis='y', labelsize=14)
    ax1.tick_params(axis='x', labelsize=14)


    t1 = np.arange(start_arrow + 20, tbif + 1, 10)
    t1 = pd.DataFrame(t1) 

    ax2.scatter(t1, df_ml.iloc[:len(t1), 0:1], color='tab:blue', label='prob. for Null')  # Adjusted y-axis to match the size of t1
    ax2.scatter(t1, df_ml.iloc[:len(t1), 1:2], color='tab:orange', label='prob. for Transcritical')  # Adjusted y-axis to match the size of t1

    ax2.legend(fontsize = 14)
    ax2.set_ylabel('Probability', fontsize = 16)
    ax2.set_xlabel('time (days)', fontsize = 16)
    ax2.tick_params(axis='y', labelsize=14)
    ax2.tick_params(axis='x', labelsize=14)
    
    plt.tight_layout()
    plt.subplots_adjust(hspace=0)
    plt.savefig('../data/figures/SI_figures/{}_forced_plot{}.png'.format(model, i))

```
